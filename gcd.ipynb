Find the gcd of 2 numbers

```python
def gcd(a,b):
  while b:                # Loop runs until b becomes 0
    a,b=b,a % b            # Swap: a becomes b, b becomes remainder
  return a               # When b=0, return a (the GCD)

num1=int(input("enter a number"))
num2=int(input("enter a number"))
result=gcd(num1,num2)
print("GCD:",result)
```
