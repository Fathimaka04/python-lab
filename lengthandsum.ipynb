```python
l1=[int(i) for i in input("enter the list:").split()]
l2=[int(i) for i in input("enter the list:").split()]
#length
if len(l1)==len(l2):
  print("length is same")
else:
  print(" length not same")

#sum
if sum(l1)==sum(l2):
  print("sum is equal")
else:
  print("sum not equal")

#common
c=set(l1).intersection(set(l2))
if len(c)!=0:
  print('common elements: ',c)
else:
  print(" no common value")

```
