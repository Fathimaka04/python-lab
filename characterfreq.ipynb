Count the number of characters (character frequency) in a string.

```python
from collections import Counter #Counter is a special class from Pythonâ€™s built-in collections module.
s=input("enter a string")
char_frequency=Counter(s) #This counts how many times each character appears.
print("character frequency:")
for char,freq in char_frequency.items():
  print(char,":",freq)
```
