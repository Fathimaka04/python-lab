 Write lambda functions to find area of square, rectangle and triangle

```python
#square
side=int(input("enter the side of the square:"))
areasquare=lambda side:side*side
print("area of the square:",areasquare(side),"\n")

#rectangle
l=int(input("enter the length:"))
b=int(input("enter the breadth:"))
arearec=lambda l,b:l*b
print("area of the rectangle:",arearec(l,b),"\n")

#triangle
base=int(input("enter the base"))
height=int(input("enter the height:"))
areatri=lambda b,h: 0.5*b*h
print("area of the rectangle:",areatri(base,height),"\n")
```
