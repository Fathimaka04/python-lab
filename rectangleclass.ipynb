create rectangle classwith attributes length and breadth and methods to find area and perimeter.comparare two rectangle objects by their area

```python
class rectangle():
  def __init__(self,l,b):
    self.l=l
    self.b=b
  def area(self):
    return self.l*self.b
  def peri(self):
    return 2*(self.l+self.b)
l=int(input("enter the lenghth of the rectangle1:"))
b=int(input("enter the breadth of the rectangle1:"))
r1=rectangle(l,b)
a1=r1.area()
print("area of rectangle1:",a1,"\n")
print("area of the rectangle1:",r1.peri(),"\n")

q=int(input("enter the length of the second rectangle 2:"))
p=int(input("enter the breadth of the second rectangle"))
r2=rectangle(p,q)
a2=r2.area()
print("area of the rectangle2:",a2)
print("perimeter of the rectangle:",r2.peri(),"\n")

if a1>a2:
  print("rectangle 1 is bigger")
elif a1<a2:
  print("rectangle 2 is bigger")
else:
  print("same area")
```
