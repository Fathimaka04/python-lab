Create aBank account with members account number ,name,type of account and balance.Write constructor and methods to deposit at the bank and withdraw an amount from the bank.

```python
class bank():
  def __init__(self):
    self.accno=int(input("enter the account number"))
    self.name=input("enter the name")
    self.acctype=input("enter account type")
    self.balance=0

  def deposite(self,depo):#deposite
    self.balance=self.balance+depo
    return self.balance

  #withdraw
  def withDraw(self,debit):
    if self.balance==0:
      print("account is empty")
    elif self.balance<debit:
      print("cannot be debitted")
    else:
      self.balance=self.balance-debit
      return self.balance

    #display
    def display(self):
      print("account number:",self.accno)
      print("Name of the holder",self.name)
      print("account type",self.acctype)
      print("balance:",self.balance,"\n")

#object creation
customer1=bank()
print("\n1.1.deposit\n2.withdraw\n3.display account details\n4.exit\n")
while True:
  a=int(input("enter the choice"))
  if a==1:
    depo=int(input("enter the amount to deposit"))
    b=customer1.deposite(depo)
    print("customer balance:",b,"\n")
  elif a==2:
    debit=int(input("enter the amount to withdraw"))
    b=customer1.withDraw(debit)
    print("customer balance",b,"\n")
  elif a==3:
    customer1.display()
  else:
    break

```
