Aim:to generate positive list of numbers from a given list of integers

```python
l=[int(i)for i in input("enter list of integers: ").split()]
p=[i for i in l if i>=0]
print("positive integers: ",p)
```

Aim:to find the sqaure of N numbers

```python
l=[int(i) for i in input("enter a list: ").split()]
s=[i*i for i in l]
print("square",s)
```

Aim: to form a list of vowels selected from a given word

```python
word=input("enter word:")
vowels="aeiouAEIOU"
vowellist=[i for i in word if i in vowels]
print("vowels in",word,"are:",vowellist)
```

Aim: to  list ordinal value of each element of a word

```python
word=input("enter a word")
ordinal_values=[ord(char)for char in word]
print(" the ordinal values of the characters are:",ordinal_values)
```
